### Formateando strings

**formateo tradicional**

```python
nombre = "Juan"
edad = 25
print("Me llamo ", nombre, " y tengo ", edad, " años.")
print("Me llamo " + str(nombre) + " y tengo " + str(edad) + " años.")
print("Me llamo %s y tengo %d años." % (nombre, edad))
```

**Usando f strings**

```python
saludos = f"Me llamo {nombre} y tengo {edad} años."
print(saludos)
```

```python
valor = 123.0
print(f"Valor: {valor:0.2f}")
```

## Mini app para calcular propinas

```python
monto = float(input("Monto de la cuenta: "))
porcentaje = float(input("Porcentaje de propina: "))
personas = int(input("Número de personas: "))

propina = monto * (porcentaje / 100)
total = monto + propina
por_persona = total / personas

print("*" * 30)
print(f"Propina: {propina:.2f}")
print(f"Total a pagar: {total:.2f}")
print(f"Cada persona paga: {por_persona:.2f}")
```
