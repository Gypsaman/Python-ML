# Matplotlib 

Este cuaderno contiene **15 ejemplos** que cubren diferentes tipos de gráficas y configuraciones típicas:
- títulos, ejes, leyendas, rejillas
- subplots (múltiples gráficas en una figura)
- anotaciones, heatmaps, bandas sombreadas, etc.

Nota: el estilo visual puede variar según tu entorno, pero los conceptos son los mismos.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

rng = np.random.default_rng(42)
plt.rcParams["figure.figsize"] = (7, 4)  # tamaño estándar para clase
```

## Ejemplo 1 — Gráfica de línea básica

**Objetivo:** graficar `sin(x)` con título y ejes.

**Conceptos clave:** `plot`, `title`, `xlabel`, `ylabel`, `grid`, `legend`.

**Qué observar:** cómo la cuadrícula ayuda a leer valores.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

plt.figure()
plt.plot(x, y, label="sin(x)")
plt.title("Gráfica de línea: sin(x)")
plt.xlabel("x (radianes)")
plt.ylabel("sin(x)")
plt.grid(True)  # cuadrícula para facilitar lectura
plt.legend()
plt.show()
```

## Ejemplo 2 — Múltiples líneas y leyenda

**Objetivo:** comparar `sin(x)` y `cos(x)` en la misma figura.

**Conceptos clave:** varias llamadas a `plot` + `legend`.

**Qué observar:** cómo la leyenda evita confusiones.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
x = np.linspace(0, 2*np.pi, 200)
plt.figure()
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.title("Múltiples líneas")
plt.xlabel("x (radianes)")
plt.ylabel("valor")
plt.legend()
plt.show()
```

## Ejemplo 3 — Scatter con datos reales (Iris)

**Objetivo:** scatter plot de dos variables del dataset Iris, coloreado por especie.

**Conceptos clave:** `scatter`, color por categoría, `colorbar`.

**Qué observar:** si hay separación visual entre clases.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
iris = load_iris(as_frame=True)
df = iris.frame.copy()
species = df["target"].to_numpy()
x = df["petal length (cm)"].to_numpy()
y = df["petal width (cm)"].to_numpy()

plt.figure()
sc = plt.scatter(x, y, c=species)
plt.title("Iris: largo de pétalo vs ancho de pétalo")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.colorbar(sc, ticks=[0, 1, 2], label="especie (0=setosa, 1=versicolor, 2=virginica)")
plt.show()
```

## Ejemplo 4 — Barras (bar chart)

**Objetivo:** comparar cantidades por categoría.

**Conceptos clave:** `bar`.

**Qué observar:** cambia valores y ve cómo se reescala el eje Y.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
categories = ["A", "B", "C", "D"]
counts = rng.integers(5, 25, size=len(categories))

plt.figure()
plt.bar(categories, counts)
plt.title("Barras: conteo por categoría")
plt.xlabel("Categoría")
plt.ylabel("Conteo")
plt.show()
```

## Ejemplo 5 — Barras horizontales

**Objetivo:** la misma idea que barras, pero horizontal.

**Conceptos clave:** `barh`.

**Qué observar:** útil cuando las etiquetas son largas.

**Qué deberías observar al ejecutar:**

## Ejemplo 6 — Histograma

**Objetivo:** visualizar la distribución de datos.

**Conceptos clave:** `hist`, `bins`.

**Qué observar:** cambia `bins` y mira cómo cambia el detalle.

## Ejemplo 7 — Boxplot

**Objetivo:** comparar dispersión entre grupos.

**Conceptos clave:** `boxplot`, mediana, cuartiles, outliers.

**Qué observar:** cuál grupo tiene mayor variabilidad.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
group1 = rng.normal(0, 1, size=200)
group2 = rng.normal(1, 1.2, size=200)
group3 = rng.normal(-0.5, 0.8, size=200)

plt.figure()
plt.boxplot([group1, group2, group3], labels=["G1", "G2", "G3"])
plt.title("Boxplot: comparación de grupos")
plt.xlabel("Grupo")
plt.ylabel("Valor")
plt.show()
```

## Ejemplo 8 — Barras de error

**Objetivo:** graficar medias con incertidumbre.

**Conceptos clave:** `errorbar`, `yerr`.

**Qué observar:** si aumentas `yerr`, las barras crecen.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
x = np.arange(5)
means = rng.normal(10, 2, size=5)
stds = rng.random(5) + 0.5

plt.figure()
plt.errorbar(x, means, yerr=stds, fmt="o")
plt.title("Barras de error: media ± desviación")
plt.xlabel("Experimento")
plt.ylabel("Medición")
plt.show()
```

## Ejemplo 9 — Subplots 1x2

**Objetivo:** poner dos gráficas en la misma figura.

**Conceptos clave:** `plt.subplots`, trabajar con objetos `ax`.

**Qué observar:** diferencia entre `plt.plot` (estado global) y `ax.plot` (orientado a objetos).

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
x = np.linspace(0, 2*np.pi, 200)
fig, ax = plt.subplots(1, 2)

ax[0].plot(x, np.sin(x))
ax[0].set_title("sin(x)")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].plot(x, np.cos(x))
ax[1].set_title("cos(x)")
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")

fig.suptitle("Dos subplots en una figura")
fig.tight_layout()
plt.show()
```

## Ejemplo 10 — Subplots 2x2

**Objetivo:** una grilla de 4 gráficas.

**Conceptos clave:** indexación `ax[r, c]`, límites `set_ylim`.

**Qué observar:** `tan(x)` crece mucho; por eso limitamos el eje Y.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
x = np.linspace(0, 2*np.pi, 200)
fig, ax = plt.subplots(2, 2)

ax[0,0].plot(x, np.sin(x)); ax[0,0].set_title("sin(x)")
ax[0,1].plot(x, np.cos(x)); ax[0,1].set_title("cos(x)")

ax[1,0].plot(x, np.tan(x)); ax[1,0].set_title("tan(x)")
ax[1,0].set_ylim(-5, 5)  # tan se dispara cerca de pi/2

ax[1,1].plot(x, np.sin(x) * np.cos(x)); ax[1,1].set_title("sin(x)cos(x)")

for a in ax.ravel():
    a.set_xlabel("x")
    a.set_ylabel("y")

fig.suptitle("Grilla 2x2 de subplots")
fig.tight_layout()
plt.show()
```

## Ejemplo 11 — Dos ejes Y (twinx)

**Objetivo:** graficar dos variables con escalas distintas en el mismo eje X.

**Conceptos clave:** `twinx`.

**Qué observar:** aunque comparten X, cada curva usa su propio eje Y.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
x = np.linspace(0, 10, 200)
y1 = np.exp(-x/3)   # decaimiento
y2 = np.sin(x)      # oscilación

fig, ax1 = plt.subplots()
ax1.plot(x, y1, label="exp(-x/3)")
ax1.set_xlabel("x")
ax1.set_ylabel("decaimiento")

ax2 = ax1.twinx()
ax2.plot(x, y2, label="sin(x)")
ax2.set_ylabel("sin")

plt.title("Ejemplo de ejes Y dobles (twinx)")
fig.tight_layout()
plt.show()
```

## Ejemplo 12 — Anotar un punto

**Objetivo:** señalar un punto relevante (por ejemplo, máximo).

**Conceptos clave:** `annotate` y `arrowprops`.

**Qué observar:** la anotación hace la gráfica más “explicativa” para presentaciones.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

imax = np.argmax(y)          # índice del máximo
xmax, ymax = x[imax], y[imax]

plt.figure()
plt.plot(x, y)
plt.title("Anotación del máximo de sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.scatter([xmax], [ymax])
plt.annotate("máximo",
             xy=(xmax, ymax),
             xytext=(xmax+0.5, ymax-0.3),
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

## Ejemplo 13 — Heatmap (imshow)

**Objetivo:** visualizar una matriz como mapa de calor.

**Conceptos clave:** `imshow`, `colorbar`, `aspect`.

**Qué observar:** la barra de color mapea valores a intensidades.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
data = rng.normal(0, 1, size=(20, 30))

plt.figure()
img = plt.imshow(data, aspect="auto")
plt.title("Heatmap (imshow)")
plt.xlabel("Índice de columna")
plt.ylabel("Índice de fila")
plt.colorbar(img, label="valor")
plt.show()
```

## Ejemplo 14 — Banda sombreada (fill_between)

**Objetivo:** mostrar un rango alrededor de una curva (p. ej. intervalo de confianza).

**Conceptos clave:** `fill_between`, `alpha` (transparencia).

**Qué observar:** cambia `alpha` y mira la visibilidad de la banda.

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
x = np.linspace(0, 10, 200)
y = np.sin(x)
band = 0.2 + 0.1*np.cos(x)   # ancho variable de la banda

plt.figure()
plt.plot(x, y, label="señal")
plt.fill_between(x, y - band, y + band, alpha=0.3, label="banda")
plt.title("fill_between: banda sombreada")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Ejemplo 15 — Scatter con tamaño variable

**Objetivo:** codificar una tercera variable usando el tamaño del marcador.

**Conceptos clave:** parámetro `s` en `scatter`.

**Qué observar:** tamaños mayores representan puntos más “lejanos” del origen (en este ejemplo).

**Qué deberías observar al ejecutar:**
- La gráfica resultante.
- El impacto de títulos, etiquetas, leyenda y límites.


```python
n = 200
x = rng.normal(0, 1, size=n)
y = rng.normal(0, 1, size=n)
sizes = (np.abs(x) + np.abs(y)) * 50   # tamaño proporcional a distancia aproximada

plt.figure()
plt.scatter(x, y, s=sizes, alpha=0.6)
plt.title("Scatter con tamaños variables")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
