## Tensores, Autograd y Grafos Computacionales

````python
import torch
````

### Tensores básicos

````python
x = torch.tensor([1.0, 2.0, 3.0])
x

x.shape, x.dtype
````

### Tensores con gradientes

````python
x = torch.tensor(2.0, requires_grad=True)
w = torch.tensor(3.0, requires_grad=True)
y = x * w
y
````

### Backward pass

````python
y.backward()
x.grad, w.grad
````

### Función más compleja

````python
x = torch.tensor(2.0, requires_grad=True)
y = x**2 + 3*x + 1
y.backward()
x.grad
````

### Tensores vectoriales

````python
v = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
out = v.sum()
out.backward()
v.grad
````

### Grafo computacional explícito


````python
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)
c = a * b
d = c + a
d.backward()
a.grad, b.grad
````

```text
a ----┐  
      ├─ (*) ── c ── (+) ── d  
b ----┘            ↑  
                   |  
                   a  
```

d = a · b + a

Entonces:

∂d / ∂a = b + 1  
∂d / ∂b = a

Sustituyendo valores:

a = 2  
b = 3

∂d / ∂a = 3 + 1 = 4  
∂d / ∂b = 2


### Reset de gradientes

````python
a.grad.zero_()
b.grad.zero_()
print(a.grad, b.grad)
````

## Conclusión
Autograd es la base de backpropagation y del entrenamiento de redes neuronales.
